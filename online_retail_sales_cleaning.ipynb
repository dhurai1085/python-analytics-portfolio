```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/dhurai1085/python-analytics-portfolio/refs/heads/main/raw_dataset/online_retail_sales_data.csv'
df= pd.read_csv(url)
df_copy = df.copy()
df_copy.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['Raw_Weight'].tail(20)
```

```python
df['Raw_Weight'] = df['Raw_Weight'].str.extract('(\d+)')
```

```python
df['Raw_Weight'] = pd.to_numeric(df['Raw_Weight'],errors='coerce')
```

```python
df['Raw_Weight'].isnull().sum()
```

```python
df['Raw_Weight'].dtype
```

```python
df['Raw_Weight'] = df['Raw_Weight'].fillna(df['Raw_Weight'].median())
df['Raw_Weight'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['ProductName'])
```

```python
df.loc[:,'Brand'] = df['Brand'].fillna('Unknown')
```

```python
df.loc[:,'UnitPrice'] = df['UnitPrice'].fillna(df['UnitPrice'].median())
```

```python
df.isnull().sum()
```

```python
# checking for duplicates

df.duplicated().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df = df.rename(columns={
    'orderid' : 'order_id',
    'customerid': 'customer_id',
    'productname': 'product_name',
    'orderdate' : 'order_date',
    'unitprice': 'unit_price'

})
```

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```
